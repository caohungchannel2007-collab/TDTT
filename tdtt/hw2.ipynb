```python
import math
R = float(input("bán kính:"))
Ix, Iy = map(float,input("nhập tọa độ tâm x, y ").split())
if R>0:
    a,b = map(float,input("nhập vào tọa độ x, y của A:").split())
    if math.sqrt((a-Ix)**2 + (b-Iy)**2) == R:
     print("A có thuộc đường tròn ")
    else:
       print("A không thuộc đường tròn")
else:
   print(" nhập lại R")
```

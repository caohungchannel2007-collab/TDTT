Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số

```python
def Amstronhg(n):
    so = n
    total = 0
    x = len(str(n))
    while n !=0:
        total = (n%10)**x + total
        n //=10
    if total == so:
        return True
    else:
        return False
```

W5A22 Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.


```python
def sum_halves(s):
    # Tách 3 chữ số đầu (từ chỉ mục 0 đến 2)
    first_half = int(s[:3])
    # Tách 3 chữ số cuối (từ chỉ mục 3 đến 5)
    second_half = int(s[3:])
    return first_half + second_half
s = input()
print(sum_halves(s))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (x2 - x1) * (y3 - y1) == (y2 - y1) * (x3 - x1)
x1, y1, x2, y2, x3, y3 = map(int,input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),


```python
def point_in_quadrilateral(quad, p):
    def cross_product(a, b, c):
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])
    n = len(quad)
    has_positive = False
    has_negative = False
    for i in range(n):
        p1 = quad[i]
        p2 = quad[(i + 1) % n]  
        cp = cross_product(p1, p2, p)
        if cp > 0:
            has_positive = True
        elif cp < 0:
            has_negative = True
        if has_positive and has_negative:
            return False
    return True
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    x12 = x2 - x1
    x13 = x3 - x1
    y12 = y2 - y1
    y13 = y3 - y1
    x23 = x3 - x2
    y23 = y3 - y2
    ab = (x12**2 + y12**2)**0.5
    ac = (x13**2 + y13**2)**0.5
    cb = (x23**2 + y23**2)**0.5
    cosgoc = (ab**2+cb**2-ac**2)/(2*ab*cb)
    cosgoc1 = math.degrees(math.acos(cosgoc))
    return cosgoc1
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(goc(x1,y1,x2,y2,x3,y3))
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]

```python
def get_max_min(a, b, c):
    max_val = max(a, b, c)
    min_val = min(a, b, c)
    return [max_val, min_val]
a, b, c = map(int,input().split())
print(get_max_min(a, b, c))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 


```python
def sum(n):
    total = 0
    for i in range(n+1):
        if len(set(str(i))) == len(str(i)):
            total += i
    return total
n = int(input())
print(sum(n))
```

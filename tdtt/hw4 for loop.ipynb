W4A8 Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def reverse(n):
    opposite = 0 
    while n!=0:
        opposite = opposite*10 +n%10
        n//=10
    return opposite
n = int(input())
if n == reverse(n):
    print(f"step = 0, palindrome = {n}")
else:
    step = 0
    while True:
        n = n + reverse(n)
        step +=1
        if n == reverse(n):
            print(f"step = {step}, palindrome = {n}")
            break
```

W4A10 Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất

```python
def collatz(i): 
    step = 1 
    while i!=1 :     
        if i %2==0:
            i//=2
        else :
            i=3*i+1
        step+=1
    return step
n = int(input())
max_step = 1
best_step = 1
for i in range(1,n+1):
    distance = collatz(i)
    if distance > max_step:
        max_step = distance
        best_step = i
print(max_step, best_step)

```

Ví dụ 220 và 284 là một cặp số thân thiết

Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
def uocso(n):
    total = 0
    for i in range(1,n):
        if n%i==0:
            total+=i
    return total
x = int(input())
y = int(input())
if uocso(x) == y and uocso(y) == x:
 print("True")
else:
   print("False")

```

W4A15 "Vừa gà vừa chó,
Bó lại cho tròn,
Ba mươi sáu con,
Một trăm chân chẵn"
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".

```python
number_of_legs = int(input())
total_animals = int(input())
for chickens in range(0, total_animals):
    dogs = total_animals - chickens
    if (dogs*4 + chickens*2==number_of_legs):
        print(f"chickens = {chickens}, dogs = {dogs}")
        break
else:
    print("invalid")
```

W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def primenumber(n):
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
        return True
primenumber(8)
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def perfect(n):
    total = 0
    n2 = n
    for i in range(1,n2):
        if n2 % i == 0:
            total +=i
    if total == n:
        return True
    else:
        return False
n = int(input())
print(perfect(n))
```

```python
def find_index(lit, k):
    for i in range(len(lit)):
            if int(lit[i]) == k:
                return i+1
    return -1
z = list(map(int, input().split()))
k = int(input())
print(find_index(z, k))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def giai_thua(n):
    x = 1
    for i in range(1,n+1):
        x *=i
    return x
n = int(input())
print(giai_thua(n))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def total_number(n):
    total = 0
    while n!=0:
        total +=n%10
        n //= 10
    return total
n = int(input())
print(total_number(n))
```

WEA12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 

```python
import math
def divide_with_remainder(a,b):
    q = a//b
    r = a-(b*q)
    if 0<= math.fabs(r)< math.fabs(q):
        return q, r
    return "invalid"
a, b = map(int,input().split())
print(divide_with_remainder(a,b))
    
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if 0<a and a<b and 0<c:
        if a**2 == b**2 +c**2 or b**2 == c**2 + a**2 or c**2 == b**2 + a**2:
            return True
    return False
a, b, c = map(int,input().split())
print(is_right_triangle(a, b, c))
```

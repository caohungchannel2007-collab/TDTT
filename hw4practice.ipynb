W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
while n!=0:
    if (n%10)%2==0:
        chan+=1
    else:
        le+=1
    n//=10
print(f"chan = {chan}, le = {le}")
```

W4A24
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A


```python
A = int(input())
total =0
step =1
while total <= A:
    total += 1/(step)
    step +=1
print(step-1)
```

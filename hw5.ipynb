Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số

```python
def Amstronhg(n):
    so = n
    total = 0
    x = len(str(n))
    while n !=0:
        total = (n%10)**x + total
        n //=10
    if total == so:
        return True
    else:
        return False
```

```python
def sochinhphuong(n):
    if int(n**0.5) == n**0.5:
        return True
    else:
        return False
sochinhphuong(16)
```

```python
def palindrome(n):
    if n == int(str(n)[::-1]):
        return True
    else:
        return False
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    if s == int(s):
        return int
    elif s==float(s):
        return float
    else:
        return "neither"
detect_number_type(12)
```
